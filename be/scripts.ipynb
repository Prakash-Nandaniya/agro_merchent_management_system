# connecting  VM to terminal via SSH keys

### create SSH key in local machine
- ssh-keygen -t rsa -b 4096

### see public key and then copy it
- cat ~/.ssh/id_rsa.pub   
- type .ssh\id_rsa.pub  -> use this command for windows

### in VM instance click edit, scroll and serch for SSH key, 
### if wants to remove old access of device delete key or now just add new SSH key and past this public key we just copied.
### hit save.

### in VM instance there will be external ip remember it.
### inside yout VM instance there will section of SSH keys look for username of your key.
- ssh USERNAME@YOUR_EXTERNAL_IP     
### run this command in device terminal we are connected now we are in local root of VM device 

### to change env variables run this command and edit file and then hit (Ctr+O, ENTER, Ctr+X)
- nano .env

### now stop old container delete container, this doesnt delete image ok just container sandbox is deleted.
- sudo docker stop CONTAINER_NAME
- sudo docker rm CONTAINER_NAME

### start container again same or new depends on you
- sudo docker run -d -p 80:80 --name CONTAINER_NAME --restart unless-stopped --env-file .env prakashnandaniya/karmatrading:v1

### to see whts gogin on in VM regarding docker use this commands.
- sudo docker ps -a   ->list all containers stopped or running.
- sudo docker images  ->list all images


### always do local development inside container. 
### This is basic goal of container to avoid problem that is running in my machine.
### to live connect code and container we use docker-compose.yml file 
- docker compose up -d   -> this will strat container with live update in code.
- docker compose down    -> to kill testing container

### when code changed and wants to reflect on server then build new image and push it but chnage version.
- docker build -t prakashnandaniya/karmatrading:v2 .
- docker push prakashnandaniya/karmatrading:v2
